```python
def plt(n, **kwargs):
    likes = np.load('likes10M_n%d.npy' % n)
    #print likes
    ks = likes[:,0]
    likes = likes[:,1]
    bics = np.array([ - 2. * likes + c * ks ** 2  for c in np.logspace(0, 4, 50) ])
    boms = np.array([ ks[np.argmin(b)] for b in bics ])
    plot(np.logspace(0,4,50), boms, **kwargs)
    ylim(0)
    xscale('log')
    xlabel('Over-Fitting Coefficient')
    ylabel('Optimal Model (k)')
```

```python
def plt_acf(n, **kwargs):
    try:
        ind = kwargs.pop('ind')
    except:
        ind = 50
    As = np.array([ np.load('MSMs/k%d_n%d/auto_rmsd.npy' % (k, n))[:,1] for k in range(50, 1050, 50) ])
    s = As[:, :ind]
    
    err = np.abs((s[1:] - s[:-1])).sum(axis=1)
    
    plot(ks[2:], err, **kwargs)
```

```python
for i in range(2, 7):
    plt_acf(i, lw=5, alpha=0.7, color=COLORS[i-2], label='$\mathrm{\mathbb{R}}^%d$' % i)

ylabel('Error in ACF')
yticks([])
xticks(np.arange(0,1250, 250))
xlim(0)
legend(prop={'size' : 16})
xlabel('Number of States')
```
