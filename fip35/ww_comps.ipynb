I want to correlate these BOMs with the autocorrelation functions predicted by each model.

```python
As_n = [ np.array([ np.load('MSMs/k%d_n%d/auto_rmsd.npy' % (k, n) )[:,1] for k in range(50, 1050, 50) ]) for n in range(2,7) ]
# As_n is a list of autocorrelation functions for each value of n

a_ref = np.load('auto_RMSD.npy')[::250][:,2]
i = np.where(a_ref <= 0)[0][0]

a_ref = a_ref[:i]

#i = np.min([ np.where(r <= 0)[0][0] for r in As ] + [ np.where(a_ref <= 0)[0][0]] )

As_n = np.array([ np.array([ r[:i] for r in As_n[n] ]) for n in range(5) ])
    
#Cs_n = np.array([ np.array([ np.corrcoef(r, a_ref)[0,1] for r in As_n[n] ]) for n in range(5) ])

Cs_n = np.array([ np.array([ np.abs(r - a_ref).mean() for r in As_n[n] ]) for n in range(5) ])

#Cs_n = np.load('corrcoeffs.npy')

np.save('absdiffs.npy', Cs_n)
print Cs_n.min(), Cs_n.max()
```

```python
figure(figsize=(4,15))
for i in range(5):
    subplot(511 + i)
    plt(ks[i][1:], Cs_n[i], lw=3, label='n=%d' %(i+2))
    ylabel('ACF Error', color='blue')
    yticks([])
    ylim(0,0.5)
    xticks([])
    xlim(0,1000)
    if i == 4:
        xlabel('Number of States')
    
    text(600, 0.415, 'd.o.f.=%d' % ns[i])
    
    twinx()
    plt(ks[i][1:], aics[i][1:], lw=3, color='red')
    xticks([])
    yticks([])
    ylim(-9E6, -4E6)


    bom = ks[i][np.argmin(aics[i])]
    y0, y1 = ylim()
    vlines(bom, y0, y1 , color='black', linestyle='dashed', lw=3)
    xlim(0,1000)
    ylabel('AIC', color='red')
    
    
    
xticks(np.arange(0, 1250, 250))

subplots_adjust(hspace=0.03)
```

Notice that there is a big drop after ~ 300 states in the AIC's I think this might be due to miscalculation of volumes. This only happens in the higher dimensional cases (n=5 or 6)
