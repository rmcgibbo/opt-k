```python
likes = [ np.load('likes_n%d.npy' % i) for i in range(2,7) ]
ks = np.array([ l[:,0] for l in likes ])
likes = np.array([ l[:,1] for l in likes ])
aics = - 2 * likes + 2 * ks ** 2
ns = range(2,7)
num_samp = 1E6 # number of frames in WW domain data... Not obvious if this is the right number.
```

```python
def plt(x,y,**kwargs):
    plot(x, y, **kwargs)
    
```

```python
for i in range(len(ns)):
    plt(ks[i], likes[i], label='n=%d' % ns[i], lw=3)
    
yticks([])
ylabel('Log Likelihood')
xlabel('Number of States')
legend(loc=4,prop={'size':16})
```

```python
for i in range(len(ns)):
    plt(ks[i], aics[i], label='n=%d' % ns[i], lw=3)

yticks([])
ylabel('AIC')
xlabel('Number of States')
legend(prop={'size':16})
```
