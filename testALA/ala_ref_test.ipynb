```python
from msmbuilder import Trajectory
Ns = np.array([100, 1000, 5000, 10000, 20000, 50000, 100000, 500000, 1000000])
cmplx = np.array([ np.load('ComplexDih/volumes%d.npy' % n) for n in Ns ])
angle = np.array([ np.load('PhiPsi/volumes%d.npy' % n) for n in Ns ])
```

```python
for i in range(len(Ns)):
    figure()
    plot(cmplx[i], label='Complex')
    plot(angle[i], label='Phi Psi')
    title('N=%d'%Ns[i])
    legend()
```

```python
ch_c = cmplx[1:] - cmplx[:-1]
figure()
for i in range(len(Ns)-1):
    figure()
    plot(ch_c[i])
    ylim([-2,2])
    title('N=%d'%Ns[i])
```
