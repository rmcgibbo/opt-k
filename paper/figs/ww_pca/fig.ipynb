```python
import pandas
matplotlib.rcParams.update(pandas.tools.plotting.mpl_stylesheet)
matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['axes.labelcolor'] = 'black'
matplotlib.rcParams['xtick.color'] = 'black'
matplotlib.rcParams['ytick.color'] = 'black'

from mdtraj import io
```

```python
f = io.loadh('times.h5')
evals = f['vals']
eval_ks = f['ks']
print eval_ks
likes = np.loadtxt('likesk.dat')[:, 1] * -2
cross = np.loadtxt('likesk_cross.dat')[:, 1] * -2
ks = np.loadtxt('likesk_cross.dat')[:, 0]

ind = np.where((1 - np.isinf(likes)) * (1 - np.isinf(cross)))[0]
likes = likes[ind]
cross = cross[ind]
ks = ks[ind]

xl = (0, 1750)
#ks = np.array(range(25, 500, 25) + range(500, xl[1], 500))
#evals = evals[:len(ks)]
#likes = likes[:len(ks)]
bic = likes + np.log(500000 - 250.) * ks**2
aic = likes + 2 * ks ** 2

times = - 0.050 / np.log(evals)
```

```python
all_ks = np.array(range(25, 500, 25) + range(500, 25000, 500))
all_times = np.load('./n500000_alltimes.npy')
all_times = - 0.050 / np.log(all_times)
print all_ks.shape, all_times.shape
```

```python
figure(figsize=(15, 7))
subplot(121)

time_c = 'k'
aic_c = '#7A68A6'
bic_c = '#A60628'
cross_c = '#348ABD'

plot(ks, likes, lw=3, color='black', label='-2$\cdot$ logL (train)')
plot(ks, cross, lw=3, color=cross_c, label='-2$\cdot$ logL (test)')
plot(ks, aic, lw=3, color=aic_c, label='AIC')
plot(ks, bic, lw=3, color=bic_c, label='BIC')

legend(loc=1, prop={'size' : 16})
ylim(-5E6, 1E6)
xlim(xl)

ylabel('Selection Criterion')
yticks([])
xticks(np.arange(0, 2000, 500))
xlabel('Number of States')
yl = ylim()
vlines(ks[np.argmin(cross)], yl[0], yl[1], lw=3, linestyle='dashed', color=cross_c)
vlines(ks[np.argmin(bic)], yl[0], yl[1], lw=3, linestyle='dashed', color=bic_c)

subplot(122)
for i in range(9):
    plot(eval_ks, times[:, i], lw=3, color=time_c)
    
xlim(xl)
ylabel('Timescales ($\mu$s)')

ax = gca()
ax.yaxis.set_label_position('right')
ax.yaxis.set_ticks_position('right')
yticks(np.arange(0, 2500, 500) / 1000.)
xticks(np.arange(0,2000, 500))
xlabel('Number of States')
yl = ylim()

vlines(ks[np.argmin(cross)], yl[0], yl[1], lw=3, linestyle='dashed', color=cross_c)
vlines(ks[np.argmin(bic)], yl[0], yl[1], lw=3, linestyle='dashed', color=bic_c)


axes((0.64, 0.3, 0.17, 0.3), axisbg='white')
for i in xrange(9):
    plot(all_ks, all_times[:, i], color=time_c, lw=3)
    
ax = gca()
ax.yaxis.set_label_position('right')
ax.yaxis.set_ticks_position('right')
#ylabel('Timescales ($\mu$s)')
yticks([0.5, 1.5, 2.5])
xticks([0, 12500, 25000])
subplots_adjust(bottom=0.2, right=0.87, left=0.10)
savefig('ww_pca.pdf')
```
