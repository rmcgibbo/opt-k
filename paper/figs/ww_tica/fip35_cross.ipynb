```python

import pandas
print pandas.__version__
matplotlib.rcParams.update(pandas.tools.plotting.mpl_stylesheet)

matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['axes.labelcolor'] = 'black'
matplotlib.rcParams['xtick.color'] = 'black'
matplotlib.rcParams['ytick.color'] = 'black'
from mdtraj import io
```

```python
def plt_base(ks, likes, cross, ts, vol=True):

    params = ks * (ks - 1) / 2.
    aic = - 2 * likes + 2 * params
    bic = - 2 * likes + np.log(1E6) * params

    
    figure(figsize=(15, 7))
    subplot(121)

    cross_c = '#348ABD'
    aic_c = '#7A68A6'
    bic_c = '#A60628'
    fill_c = 'gray'
    m = None
    plot(ks, -2 * likes, marker=m, lw=3, color='black', label='-2$\cdot$ logL (train)')
    if not cross is None:
        plot(ks, -2 * cross, marker=m, lw=3, color=cross_c, label='-2$\cdot$ logL (test)')
    plot(ks, aic, marker=m, lw=3, color=aic_c, label='AIC')
    plot(ks, bic, marker=m, lw=3, color=bic_c, label='BIC')
    
    if not cross is None:
        d = np.max(np.concatenate([aic, -2 * likes, -2 *cross])) - (-2 * likes).min()
    else:
        d = np.max(np.concatenate([aic, -2 * likes])) - (-2 * likes).min()
    
    ymin = np.min(-2 * likes) 
    ymax = ymin + 2 * d
    yl = (ymin, ymax)
    vlines(ks[aic.argmin()], yl[0], yl[1], color=aic_c, linestyle='dashed', lw=3)
    vlines(ks[bic.argmin()], yl[0], yl[1], color=bic_c, linestyle='dashed', lw=3)
    if not cross is None:
        vlines(ks[cross.argmax()], yl[0], yl[1], color=cross_c, linestyle='dashed', lw=3)
        f0 = ks[np.min([aic.argmin(), bic.argmin(), cross.argmax()])]
        f1 = ks[np.max([aic.argmin(), bic.argmin(), cross.argmax()])]
    else:
        f0 = ks[np.min([aic.argmin(), bic.argmin()])]
        f1 = ks[np.max([aic.argmin(), bic.argmin()])]
    
    fill_betweenx(yl, f0, f1, color=fill_c, alpha=0.2)
    ylim(yl)
    xlabel('Number of States')
    ylabel('Selection Criterion')
    yticks([])
    legend(loc=1, prop={'size' : 16})
    xlim(0, 1000)
    subplot(122)

    for i in range(10):
        plot(ks, ts[:, i], lw=3, color='black', marker=m)
    

    xlabel('Number of States')
    ylabel('Timescales ($\mu$s)')
    #yscale('log')

    yl = (0, 6)
    vlines(ks[aic.argmin()], yl[0], yl[1], color=aic_c, linestyle='dashed', label='AIC Opt.', lw=3)
    vlines(ks[bic.argmin()], yl[0], yl[1], color=bic_c, linestyle='dashed', label='BIC Opt.', lw=3)
    if not cross is None:
        vlines(ks[cross.argmax()], yl[0], yl[1], color=cross_c, linestyle='dashed', label='Cross Opt.', lw=3)
    fill_betweenx(yl, f0, f1, color=fill_c, alpha=0.2)
    ylim(yl)
    xlim(0, 1000)
    #legend(loc=1, prop={'size' : 18})
    ax = gca()
    ax.yaxis.set_ticks_position('right')
    ax.yaxis.set_label_position('right')
    
    subplots_adjust(bottom=0.2, right=0.87, left=0.10)
    
```

```python
v = io.loadh('volK.h5')
likeV = v['likes']
crossV = v['cross']
ks = v['ks']

#g = io.loadh('mvnK.h5')
#likeG = g['likes']
#crossG = g['cross']
#ks = g['ks']

valsK = io.loadh('evalsK.h5', 'evals')
timesK = -0.050 / np.log(valsK)

plt_base(ks, likeV, crossV, timesK, vol=True)
savefig('ww_tica.pdf')
#plt_base(ks, likeG, crossG, timesK, vol=False)
#savefig('ww_likeKmvn.pdf')
```

```python
v = io.loadh('volNP.h5')
likeV = v['likes']
crossV = v['cross']
ks = v['ks']

g = io.loadh('mvnNP.h5')
likeG = g['likes']
crossG = g['cross']
ks = g['ks']

vals = io.loadh('evalsNP.h5', 'evals')
times = -50 / np.log(vals)

plt_base(ks, likeV, cross=None, ts=times, vol=True)
savefig('ww_likeNPvol.pdf')
plt_base(ks, likeG, cross=None, ts=times, vol=False)
savefig('ww_likeNPmvn.pdf')
```
