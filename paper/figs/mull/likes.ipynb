```python
def plt(mvn=False, x_lbl=True):
    
    if mvn:
        l = np.load('mvn/MSMs_train_k.npy')[1:]
        c = np.load('mvn/MSMs_k.npy')[1:]
    else:
        l = np.load('MSMs_train_k.npy')[1:]
        c = np.load('MSMs_k.npy')[1:]
    
    n = np.concatenate([np.arange(50, 1050, 50), np.arange(2000, 4000, 1000)])
    ts = np.array([np.load('MSMs/k%d/Lag30k/times10.npy' % i) for i in n[:]])

    bic = - 2 * l + n ** 2 * log(1E6)
    aic = - 2 * l + n ** 2 * 2
    plot(n, -2 * l, lw=3, color='k', label='-2$\cdot$ logL (train)')
    plot(n, -2 * c, lw=3, label='-2$\cdot$ logL (test)')
    plot(n, aic, lw=3, label='AIC')
    plot(n, bic, lw=3, label='BIC')

    ymin = np.min(-2 * l)
    ymax = ymin + (bic[0] - ymin) * 2

    xlim(0, 2000)
    xt_locs = np.arange(0, 2500, 500)
    if x_lbl:
        xticks(xt_locs)
        xlabel('Number of States')
    else:
        xticks(xt_locs, [])
        
    yticks([])
    ylabel('Score (%s)' % ('MVN' if mvn else 'Volume'))
    ylim(ymin, ymax)
    ax = gca()
    ax.yaxis.set_label_coords(0, 0.5)
    legend(loc=1, prop={'size' : 18})
    
def plt_time(right=False, x_lbl=True):
    for i in xrange(2):
        plot(n, -30 / np.log(ts[:, i]), lw=3)
    

    yscale('log')
    yticks(yticks()[0], [])
    ylim([50, 1E4])
    ylabel('Timescale')
    ax = gca()
    ax.yaxis.set_label_coords(0, 0.5)

    xlim(0, 2000)
    xt = np.arange(0, 2500, 500)
    
    if x_lbl:
        xticks(xt)
        xlabel('Number of States')
    else:
        xticks(xt, [])
    
    if right:
        ax=gca()
        ax.yaxis.set_label_position('right')
        ax.yaxis.tick_right()

figure(figsize=(7, 12))
subplot(311)
plt(False, x_lbl=False)
subplot(312)
plt(True, x_lbl=False)
subplot(313)
plt_time(x_lbl=True)

subplots_adjust(hspace=0.03)
savefig('mull_cross_vert.png', dpi=200)
```
