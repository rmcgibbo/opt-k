```python

import pandas
print pandas.__version__
matplotlib.rcParams.update(pandas.tools.plotting.mpl_stylesheet)

matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['font.family'] = 'sans-serif'
matplotlib.rcParams['axes.labelcolor'] = 'black'
matplotlib.rcParams['xtick.color'] = 'black'
matplotlib.rcParams['ytick.color'] = 'black'
from mdtraj import io
```

```python
def plt_base(ks, likes, cross, ts, vol=True):
    aic = - 2 * likes + 2 * ks**2
    bic = - 2 * likes + np.log(1E6) * ks**2

    
    figure(figsize=(15, 7))
    subplot(121)

    cross_c = '#348ABD'
    aic_c = '#7A68A6'
    bic_c = '#A60628'
    fill_c = 'gray'
    m = None
    plot(ks, -2 * likes, marker=m, lw=3, color='black', label='-2$\cdot$ logL (train)')
    plot(ks, -2 * cross, marker=m, lw=3, color=cross_c, label='-2$\cdot$ logL (test)')
    plot(ks, aic, marker=m, lw=3, color=aic_c, label='AIC')
    plot(ks, bic, marker=m, lw=3, color=bic_c, label='BIC')
    ymin = np.min(-2 * likes)
    ymax = ymin + (bic[0] - ymin) * 2
    yl = (ymin, ymax)
    vlines(ks[aic.argmin()], yl[0], yl[1], color=aic_c, linestyle='dashed', lw=3)
    vlines(ks[bic.argmin()], yl[0], yl[1], color=bic_c, linestyle='dashed', lw=3)
    vlines(ks[cross.argmax()], yl[0], yl[1], color=cross_c, linestyle='dashed', lw=3)
    f0 = ks[np.min([aic.argmin(), bic.argmin(), cross.argmax()])]
    f1 = ks[np.max([aic.argmin(), bic.argmin(), cross.argmax()])]
    
    fill_betweenx(yl, f0, f1, color=fill_c, alpha=0.2)
    ylim(yl)
    xlabel('Number of States')
    ylabel('Selection Criterion')
    yticks([])
    legend(loc=1, prop={'size' : 16})
    xlim(0, 1000)
    subplot(122)

    for i in range(5):
        plot(ks, ts[:, i], lw=3, color='black', marker=m)
    

    xlabel('Number of States')
    ylabel('Timescales')
    gca().yaxis.set_ticklabels([])
    yscale('log')

    yl = (20, 1E4)
    vlines(ks[aic.argmin()], yl[0], yl[1], color=aic_c, linestyle='dashed', label='AIC Opt.', lw=3)
    vlines(ks[bic.argmin()], yl[0], yl[1], color=bic_c, linestyle='dashed', label='BIC Opt.', lw=3)
    vlines(ks[cross.argmax()], yl[0], yl[1], color=cross_c, linestyle='dashed', label='Cross Opt.', lw=3)
    print cross.argmin()
    fill_betweenx(yl, f0, f1, color=fill_c, alpha=0.2)
    ylim(yl)
    xlim(0, 1000)
    #legend(loc=1, prop={'size' : 18})
    ax = gca()
    ax.yaxis.set_ticks_position('right')
    ax.yaxis.set_label_position('right')
    
    subplots_adjust(bottom=0.2, right=0.87, left=0.10)
    
```

```python
f = io.loadh('timescales.h5')
ts = - 30 / np.log(f['timescales'])
n = f['num_states']
#l_mvn = np.load('mvn/MSMs_train_k.npy')[1:]
#c_mvn = np.load('mvn/MSMs_k.npy')[1:]
l_vol = np.load('MSMs_train_k.npy')[1:]
c_vol = np.load('MSMs_k.npy')[1:]

#plt_base(n, l_mvn, c_mvn, ts, vol=False)
#savefig('mull_mvn.pdf')
plt_base(n, l_vol, c_vol, ts, vol=True)
savefig('mull.pdf')
```

```python
def plt(mvn=False, x_lbl=True):
    
    if mvn:
        l = np.load('mvn/MSMs_train_k.npy')[1:]
        c = np.load('mvn/MSMs_k.npy')[1:]
    else:
        l = np.load('MSMs_train_k.npy')[1:]
        c = np.load('MSMs_k.npy')[1:]
    
    #n = np.concatenate([np.arange(50, 1050, 50), np.arange(2000, 4000, 1000)])
    #ts = np.array([np.load('MSMs/k%d/Lag30k/times10.npy' % i) for i in n[:]])

    
    bic = - 2 * l + n ** 2 * log(1E6)
    aic = - 2 * l + n ** 2 * 2
    plot(n, -2 * l, lw=3, color='k', label='-2$\cdot$ logL (train)')
    plot(n, -2 * c, lw=3, label='-2$\cdot$ logL (test)')
    plot(n, aic, lw=3, label='AIC')
    plot(n, bic, lw=3, label='BIC')

    ymin = np.min(-2 * l)
    ymax = ymin + (bic[0] - ymin) * 2

    xlim(0, 2000)
    xt_locs = np.arange(0, 2500, 500)
    if x_lbl:
        xticks(xt_locs)
        xlabel('Number of States')
    else:
        xticks(xt_locs, [])
        
    yticks([])
    ylabel('Score (%s)' % ('MVN' if mvn else 'Volume'))
    ylim(ymin, ymax)
    ax = gca()
    ax.yaxis.set_label_coords(0, 0.5)
    legend(loc=1, prop={'size' : 18})
    
def plt_time(right=False, x_lbl=True):
    
    for i in xrange(2):
        plot(n, -30 / np.log(ts[:, i]), lw=3)
    

    yscale('log')
    yticks(yticks()[0], [])
    ylim([50, 1E4])
    ylabel('Timescale')
    ax = gca()
    ax.yaxis.set_label_coords(0, 0.5)

    xlim(0, 2000)
    xt = np.arange(0, 2500, 500)
    
    if x_lbl:
        xticks(xt)
        xlabel('Number of States')
    else:
        xticks(xt, [])
    
    if right:
        ax=gca()
        ax.yaxis.set_label_position('right')
        ax.yaxis.tick_right()

figure(figsize=(7, 12))
subplot(311)
plt(False, x_lbl=False)
subplot(312)
plt(True, x_lbl=False)
subplot(313)
plt_time(x_lbl=True)

subplots_adjust(hspace=0.03)
savefig('mull_cross_vert.png', dpi=200)
```
