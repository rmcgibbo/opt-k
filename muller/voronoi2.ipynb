```python
from mullermsm import muller
from pyschwancr import custom_tickers
s_fmtr = custom_tickers.SchwancrFormatter(useMathText=True)
```

```python
k10 = np.load('data/gens10.npy')
k50 = np.load('data/gens50.npy')
k100 = np.load('data/gens100.npy')
k250 = np.load('data/gens250.npy')
k800 = np.load('data/gens800.npy')
h100 = np.load('data/gens100hybrid.npy')
m100 = np.load('data/gens100kmeans.npy')
klikes = np.load('data/KCenters_n100k.npy')
hlikes = np.load('data/Hybrid10.npy')
mlikes = np.load('data/KMeans.npy')
```

```python
limits = [-3.3, 1.3, -1.3, 2.5]
shown_limits = [-3.3, 1.2, -1.3, 2.5] # something is wrong in plot_v I think...
my_cmap = matplotlib.cm.jet_r
vor_color = 'black'
matplotlib.rcParams['font.size'] = 22
def plot_stuff(gens, plot_points=False):
    muller.plot_v(*limits, cmap=my_cmap)
    muller.plot_voronoi(gens, limits, color=vor_color)
    if plot_points:
        scatter(gens[:,0], gens[:,1], marker='*', s=100, c='')
    xlim(shown_limits[:2])
    ylim(shown_limits[2:])
    xticks([])
    yticks([])
    
```

```python
plot_stuff(k10)
title('K-Centers (k=10)')
```

```python
plot_stuff(k50)
title('K-Centers (k=50)')
text(shown_limits[0]+0.1, shown_limits[2]+0.1, '$\log L = %s$' % s_fmtr._formatSciNotation('%.2e'%klikes[1]), color='white')
```

```python
plot_stuff(k100)
title('K-Centers (k=100)')
text(shown_limits[0]+0.1, shown_limits[2]+0.1, '$\log L = %s$' % s_fmtr._formatSciNotation('%.2e'%klikes[2]), color='white')
```

```python
plot_stuff(h100)
title('Hybrid K-Medoids (k=100)')
text(shown_limits[0]+0.1, shown_limits[2]+0.1, '$\log L = %s$' % s_fmtr._formatSciNotation('%.2e'%hlikes[2]), color='white')
```

The hybrid generators are more central, and we would *expect* this model to have a higher log-likelihood!

```python
plot_stuff(m100)
title('K-Means (k=100)')
text(shown_limits[0]+0.1, shown_limits[2]+0.1, '$\log L = %s$' % s_fmtr._formatSciNotation('%.2e'%mlikes[2]), color='white')
```

The K-Means generators are even more central as compared to the Hybrid K-Medoids model. We again would expect this model to have a higher likelihood.

```python
plot_stuff(k250)
title('K-Centers (k=250)')
text(shown_limits[0]+0.1, shown_limits[2]+0.1, '$\log L = %s$' % s_fmtr._formatSciNotation('%.2e'%klikes[5]), color='white')
```
