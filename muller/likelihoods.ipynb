```python
likes = np.load('likes.npy')
ks = np.array(range(1,101) + [1000])
plot(ks, likes)
```

```python
bic = -2 * likes + ks**2 * np.log(10E7)
aic = -2 * likes + 2 * ks**2
figure()
subplot(311)
plot(ks, likes, label='Log-Likelihood')
legend(loc=4)
subplot(312)
plot(ks, bic, label='BIC')
legend()
subplot(313)
plot(ks, aic, label='AIC')
legend()
```
