```python
likes = np.load('data/likes1000_s50_n100k.npy')
#likes[np.where(likes==np.inf)] = 0
# This is because some volumes are zero, which means I need to use more 
# MC points to calculate those volumes. This data used 10,000 points
# I am calculating volumes with 100,000 points now to compare
#ks = np.array(range(1,101) + [1000])
ks = np.arange(0,1050,50)
figure()
axes( (0.15, 0.15, 0.7, 0.7) )
xlabel('Number of States')
ylabel('Log Likelihood')
ylim([0,4E7])
yticks(yticks()[0][::2])
plot(ks, likes, lw=3)
savefig('log_like.pdf')
yticks([])
savefig('log_like_noY.pdf')
```

```python
bic = -2 * likes + ks**2 * np.log(1E7)
aic = -2 * likes + 2 * ks**2
plot(ks, bic, lw=3)
xlabel('Number of States')
ylabel('BIC')
yticks([])
vlines(250, ylim()[0], ylim()[1], color='red', lw=3, linestyle='dashed', label='Optimal Model')
l=legend(prop={ 'size' : 14 })
savefig('bic_opt.pdf')
```

So the BIC does quite well at providing a "best" model. However, the question remains, is it the "best" model? One test would be to look at the slowest timescales as a function of the number of states.

```python
lams1 = np.load('data/lambda1.npy')
lams2 = np.load('data/lambda2.npy')
lams3 = np.load('data/lambda3.npy')

times1 = np.concatenate(([0], - 30 / np.log(lams1)))
times2 = np.concatenate(([0], - 30 / np.log(lams2)))
times3 = np.concatenate(([0], - 30 / np.log(lams3)))

plot(ks, times1, color='blue', lw=3)#label='$t_1$')
plot(ks, times2, color='blue', lw=3)#, label='$t_2$')
plot(ks, times3, color='blue', lw=3)# label='$t_3$')
ylabel('Timescales')
yscale('log')
xlabel('Number of States')
vlines(250,10,1E4,color='red', label='Optimal Model',lw=3, linestyle='dashed')
legend(loc=4, prop={'size' : 14})
```

```python
tlikes = np.load('data/tlikes1000_s50_n100k.npy')
vlikes = np.load('data/vlikes1000_s50_n100k.npy')
figure()
plot(ks, tlikes, label='Transition Log-Likelihood', lw=3)
tbic = - 2 * tlikes + ks ** 2 * np.log(1E7)
plot(ks, tbic, color='red', lw=3, label='Transition BIC')
plot(ks, bic, color='purple', label='Total BIC', lw=3)
legend(loc=2)
ylim(ylim()[0], 1.3E8)
None
```

If you just use the log-likelihood from the transition probabilities you just get a likelihood that decreases as the number of states increases. The BIC, therefore, does not actually peak. This is because we cannot compare the transition probability likelihoods between different state decompositions.
